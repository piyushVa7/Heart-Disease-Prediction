```python
data.groupby('TenYearCHD').mean()
```

```python
final = data.drop(['diaBP','BMI','heartRate'], axis = 1)
```

```python
heart_attack = final[final['TenYearCHD']==1]
no_heart_attack = final[final['TenYearCHD']==0]
```

```python
X = final[['male','age','cigsPerDay','BPMeds','prevalentStroke','prevalentHyp','diabetes','totChol','sysBP','glucose']]
```

```python
Y = final['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size = 0.20, random_state = 2)
```

```python
from sklearn.metrics import accuracy_score,classification_report,precision_score,recall_score
from sklearn.metrics import precision_recall_curve,confusion_matrix,roc_curve,auc,log_loss
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred_prob = model.predict_proba(X_test)[:,1]
```

```python
[fpr,tpr,thr] = roc_curve(y_test,y_pred_prob)
```

```python
print("Train test split results:")
```

```python
print(model.__class__.__name__+"accuracy is %2.3f" % accuracy_score(y_test,y_pred))
print(model.__class__.__name__+"log_loss is %2.3f" % log_loss(y_test,y_pred_prob))
print(model.__class__.__name__+'auc is %2.3f' % auc(fpr,tpr))
```
